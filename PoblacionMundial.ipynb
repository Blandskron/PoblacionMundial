```python
import matplotlib.pyplot as plt
import pandas as pd

# cargar datos de la población mundial de 1960 a 2018
url = 'https://raw.githubusercontent.com/datasets/population/master/data/population.csv'
data = pd.read_csv(url, usecols=['Country Name', 'Country Code', 'Year', 'Value'])
data = data.loc[(data['Year'] >= 1960) & (data['Year'] <= 2018)]

# crear gráfico
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(data['Year'], data['Value'], label='Población mundial')

# establecer el título y las etiquetas de los ejes
ax.set_title('Población mundial de 1960 a 2018')
ax.set_xlabel('Año')
ax.set_ylabel('Población (millones)')

# establecer límites de los ejes
ax.set_xlim(1960, 2018)
ax.set_ylim(0, max(data['Value']) * 1.1)

# agregar rejilla
ax.grid(True)

# agregar leyenda
ax.legend(loc='upper left')

# mostrar gráfico
plt.show()
```

```python
import pytz
from datetime import datetime

# Obtener la hora actual
now = datetime.now()

# Imprimir la hora actual en todas las zonas horarias GMT
for tz in pytz.all_timezones:
    if tz.startswith('Etc/GMT'):
        timezone = pytz.timezone(tz)
        dt = timezone.localize(now)
        fmt = '%Y-%m-%d %H:%M:%S %Z%z'
        print(dt.strftime(fmt))
```
